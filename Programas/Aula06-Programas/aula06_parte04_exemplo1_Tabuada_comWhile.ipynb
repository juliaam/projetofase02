```python
#-----------------------------------------------------------------------------------------------------
# LOGICA E PROGRAMACAO DE COMPUTADORES
# AULA 06 - PARTE 4
# Professora: Silvia Moraes
# Exemplo 1 : repeticao aninhada com while - Tabuada 
# ----------------------------------------------------------------------------------------------------

numero = 1
while numero<=10:
  valor = 1
  while valor<=10:
    print(numero, ' x ', valor, " = ", numero * valor)
    valor = valor + 1
  print()
  numero = numero + 1
```
