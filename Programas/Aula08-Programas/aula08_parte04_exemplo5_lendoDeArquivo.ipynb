```python
#-----------------------------------------------------------------------------------------------------
# LOGICA E PROGRAMACAO DE COMPUTADORES
# AULA 08 - PARTE 4
# Professora: Silvia Moraes
# Exemplo 5: lendo dados de um arquivo
# ----------------------------------------------------------------------------------------------------
arq = open("planta_iris.data","r")
dados = []
for linha in arq:
  if 'sepal' in linha: continue   #para pular o cabecalho
  valores = linha.split(',')
  #print(linha)
  tupla = (float(valores[0]),float(valores[1]))
  dados.append(tupla)
arq.close()
print(dados)
```
